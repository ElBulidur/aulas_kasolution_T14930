# Matplotlib

```python
df.head(2)
```

```python
X = df.Datas[0:5]
Y = df.Calories[0:5]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15,4))

ax1.plot(
    X, 
    Y,
    color="Black",
    linestyle="--",
    linewidth=4,
    marker="*"
)
ax1.set_title("Grafico de linha")
ax1.set_ylabel("Calorias")
ax1.set_xlabel("Datas")

ax2.scatter(X, Y)
ax2.set_title("Grafico de dispersao")

ax3.bar(X, Y)
ax3.set_title("Grafico de barra")


# fig.savefig("meus graficos.pdf")
plt.show()
```

```python
# varias plotagenscom PYPLOT
fig, axs = plt.subplots(3,3)

axs[2,2].plot(X, Y)
axs[1,1].set_facecolor("r")
axs[0,0].scatter(X, Y)
axs[1,1].bar(X, Y)



plt.show()
```

```python
fig, axs = plt.subplots(3, 3)

gs = fig.add_gridspec(3,3)

# Cabecalho
cab = fig.add_subplot(gs[0, :])
cab.set_facecolor("c")
cab.spines["top"].set_color("None")
cab.spines["bottom"].set_color("None")
cab.spines["left"].set_color("None")
cab.spines["right"].set_color("None")
cab.text(0.1, 0.5, "Representacao dos graficos", size=20)

# x = np.linspace(0.1, 10*np.pi, 1000)
# y = np.sin(x)

# cab.plot(x, y, "C1o", markevery=25, mec="1.0")


#https://matplotlib.org/stable/users/explain/colors/colormaps.html
cmap = plt.get_cmap("Oranges")
colors = cmap([0.2, 0.8])

# Lin
ax1 = fig.add_subplot(gs[1, 0])
ax1.plot(X, Y)

# Hist
ax1 = fig.add_subplot(gs[1, 1])
ax1.hist(df[["Calories", "Duration"]], bins=6, color=colors)

# Bar
ax1 = fig.add_subplot(gs[1, 2])
ax1.bar(df.Name, df.Calories)

# Scatter
ax1 = fig.add_subplot(gs[2, 1])
ax1.pie(df.Calories)

# fig.savefig("trabalho_apresentado.pdf")

plt.show()
```
