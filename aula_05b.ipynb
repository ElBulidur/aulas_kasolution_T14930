# PLOTAGEM

```python
import pandas as pd
import numpy as np

# mais usado
import matplotlib.pyplot as plt

# mais robustos
import matplotlib.gridspec as gridspec
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,4))

ax1.plot(
    [1,2,3,4,5,6,7],
    [1,2,1,1,4,5,7]
)
ax2.bar(
    [1,2,3,4,5,6,7],
    [1,2,1,1,4,5,7]
)

plt.show()
```

```python

x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)

plot = {
    "x":x,
    "y":y
}
df_plot = pd.DataFrame(plot)
df_plot.head()
```

```python
fig, ax = plt.subplots()

ax.plot(df_plot.x, df_plot.y)

plt.show()
```

```python
x1 = np.random.uniform(0, 1, 100)
y1 = np.random.uniform(0, 1, 100)

plot = {
    "x1":x1,
    "y1":y1
}
df_plot1 = pd.DataFrame(plot)
df_plot1.head()
```

```python
fig, ax = plt.subplots()

ax.scatter(df_plot1.x1, df_plot1.y1)

plt.show()
```

```python
x3 = np.arange(10)
y3 = np.random.uniform(1, 10, 10)

plot = {
    "x3":x3,
    "y3":y3
}
df_plot3 = pd.DataFrame(plot)
df_plot3.head()
```

```python
fig, ax = plt.subplots()

ax.bar(df_plot3.x3, df_plot3.y3)

plt.show()
```

```python
xy = np.random.uniform(0, 1, (8,8))

plot = {
    "x4":xy[0],
    "y4":xy[1]
}
df_plot4 = pd.DataFrame(plot)
df_plot4.head()
```

```python
fig, ax = plt.subplots()

ax.imshow(df_plot4, vmin=0, vmax=8)

plt.show()
```

```python
x = np.random.uniform(0, 1, 4)

s_plot1 = pd.Series(x)
s_plot1.head()
```

```python
fig, ax = plt.subplots()

labels = ["a", "b", "c", "d"]
explode = (0.1, 0.1, 0.1, 0.1)

ax.pie(
    s_plot1,
    explode=explode,
    labels=labels,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90
)

plt.show()
```
