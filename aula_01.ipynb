```python
valores = [20.9, 15.5, 30.9]
produtos = ["Manga", "Pera", "Banana"]

s2 = pd.Series(data=valores, index=produtos)
```

```python
s2.plot(figsize=(13,2))
```

```python
preco_gasolina = {
    "Semana 01": 4.99,
    "Semana 02": 9.99,
    "Semana 03": 12.90,
    "Semana 04": 14.99,
}

s3 = pd.Series(preco_gasolina)
```

```python
precos_produtos = {
    "Cerveja": 5.0,
    "Cigarro": 7.0,
    "Carvao": 21.0,
    "Gelo": 12.0,
    "Baralho": 30.0,
    "Sal Grosso": 3.0,
}

s4 = pd.Series(precos_produtos)
```

```python
s4[1:5]
```

```python
s4["Cigarro":"Baralho"]
```

```python
s4.sum()
```

```python
precos_produtos = {
    "Cerveja": [5.0, 90],
    "Cigarro": 7.0,
    "Carvao": 21.0,
    "Gelo": 12.0,
    "Baralho": 30.0,
    "Sal Grosso": 3.0,
}

s5 = pd.Series(precos_produtos, index=["Cerveja", "Gelo", "Carne"])
```

```python
s5
```

```python
s5.sum()
```

```python
s6 = pd.Series(precos_produtos, index=["Cerveja", "Gelo", "Carne", "Cerveja"])
```

```python
s6
```
